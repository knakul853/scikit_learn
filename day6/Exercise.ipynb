```python
import pandas as pd
df = pd.read_csv('salaries.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['company_n'] = le.fit_transform(df['company'])
df['job_n'] = le.fit_transform(df['job'])
df['degree_n'] = le.fit_transform(df['degree'])
df

```

```python
df = df.drop(['company','job','degree'],axis = 'columns')
df
```

```python
y = df.drop(['company_n','job_n','degree_n'],axis='columns')
x = df.drop(['salary_more_then_100k'],axis='columns')
```

```python
y
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model.fit(x,y)
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df = df.drop(['PassengerId','Name','SibSp','Parch','Cabin','Embarked','Ticket'],axis='columns')
```

```python
df.Sex = df.Sex.map({'male':1,'female':2})
```

```python
df.head()
```

```python
df.Age = df.Age.fillna(df.Age.mean())
```

```python
df.head()
X = df.drop(['Survived'],axis = 'columns')
Y = df.Survived
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train , y_train)
```

```python
model.score(x_test , y_test)
```
